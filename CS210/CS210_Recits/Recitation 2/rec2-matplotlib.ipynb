````python
import matplotlib.pyplot as plt  # we'll be using pyplot with plt alias
import numpy as np

# graphs will be included in the notebook
%matplotlib inline
````

````python
# let's start with the basics
plt.plot([1, 2, 3, 4])
plt.show()
````

````python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.show()
````

````python
# let's plot cosine and sine functions
x = np.linspace(-np.pi, np.pi, 256, endpoint=True)
c = np.cos(x)
s = np.sin(x)

plt.plot(x, c)  # first parameter x-axis, second one y-axis
plt.plot(x, s)
plt.show()
````

````python
# playing with the defaults
x = np.linspace(-np.pi, np.pi, 256, endpoint=True)
c = np.cos(x)
s = np.sin(x)

# blue color with a continuous line of width 1 (pixels)
plt.plot(x, c, color="blue", linewidth=1.0, linestyle=":", label="cosine")

# green color with a continuous line of width 1 (pixels)
plt.plot(x, s, color="green", linewidth=1.0, linestyle="-.", label="sine")

# set x limits
plt.xlim(-4.0,4.0)
# set y limits
plt.ylim(-1.0,1.0)

# set x ticks
#plt.xticks(np.linspace(-4,4,5,endpoint=True))
#plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi])
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi], [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$\pi/2$', r'$\pi$'])

# set y ticks
plt.yticks(np.linspace(-1,1,5,endpoint=True))

# enable the legend
plt.legend()

# add title
plt.title("sine - cosine")

plt.show()
````

````python
data = {
        'a': np.arange(50),
        'c': np.random.randint(0, 50, 50),
        'd': np.random.randn(50)
       }

data['b'] = data['a'] + 10 * np.random.randn(50)  # adding a new key
data['d'] = np.abs(data['d']) * 100

plt.scatter('a', 'b', c='c', s='d', data=data)  # setting x and y axes from the data
                                                # c = color coding, s = size of each dot
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()
````

````python
labels = 'Frogs', 'Hogs', 'Dogs', 'Logs'
sizes = [15, 30, 45, 10]
explode = (0, 0.1, 0, 0)  # only "explode" the 2nd slice

plt.pie(sizes, explode=explode, labels=labels, autopct='%.1f',
        shadow=True, startangle=90)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
````

````python
x = np.random.normal(10, 5, 1000)  # normal dist. with mean set to 10 and standard deviation to 5

plt.hist(x, bins=20, rwidth=0.9)
plt.show()
````

````python
x = ['Nuclear', 'Hydro', 'Gas', 'Oil', 'Coal', 'Biofuel']
energy = np.random.randint(10, 100, len(x))

x_pos = [i for i, _ in enumerate(x)]  # equivalent of np.arange(len(x))

plt.bar(x_pos, energy, color='green')
plt.xlabel("Energy Source")
plt.ylabel("Energy Output (GJ)")
plt.title("Energy output from various fuel sources")

plt.xticks(x_pos, x)

plt.show()
````

#### Playing with the layout

![](https://i.stack.imgur.com/HZWkV.png)

``` py
>>> fig = plt.figure()  # create a figure object
>>> ax = fig.add_subplot(1, 1, 1)  # create an axes object in the figure
                                   # (row, column, position)
>>> ax.plot(x)  # now, we can plot on axes ax
```

``` py
>>> fig, axs = plt.subplots(1, 3)  # returns a figure in which 3 axes are placed horizontally
```

````python
data = {'apples': 10, 'oranges': 15, 'lemons': 5, 'limes': 20}
names = list(data.keys())
values = list(data.values())

fig, axs = plt.subplots(1, 3, figsize=(10, 3), sharey=True)
axs[0].bar(names, values)
axs[1].scatter(names, values)
axs[2].plot(names, values)
fig.suptitle('Categorical Plotting')  # figure title
````

````python
x = np.linspace(-np.pi, np.pi, 256, endpoint=True)
c = np.cos(x)
s = np.sin(x)
 
plt.subplot(2,1,1)  # row, column, column position (first)
#plt.subplot(211)  # another format to represent the position
plt.title('subplot(2,1,1)')
plt.plot(x,c, 'b-.')
 
plt.subplot(2,1,2)
plt.title('subplot(2,1,2)')
plt.plot(x,s,'r-.')
 
plt.tight_layout()  # adjusting subplot positions
plt.show()
````

#### Annotations

````python
x = np.linspace(-np.pi, np.pi, 256, endpoint=True)
c = np.cos(x)
s = np.sin(x)

plt.subplots(figsize=(5, 5))

plt.plot(x, c)  # first parameter x-axis, second one y-axis
plt.plot(x, s)

plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi], [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$\pi/2$', r'$\pi$'])

t = 2*np.pi/3
plt.annotate(r'$\sin(\frac{2\pi}{3})=\frac{\sqrt{3}}{2}$',  # text
             xy=(t, np.sin(t)),                             # coordianates
             xycoords='data',                               # coordinate system
             xytext=(+10, +30),                             # text offset
             textcoords='offset points',                    # text coordinate system
             fontsize=16,                                   # font size
             arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"))  # styling for arrow 

plt.show()
````

#### Widgets

````python
from ipywidgets import interact
````

````python
x = np.linspace(-np.pi, np.pi, 256, endpoint=True)

def plotsin(f):  # parameter will be updated with the slider
    plt.plot(x, np.sin(2*x*f))
    plt.show()
    
interact(plotsin, f=(1, 10, 1))  # start, end, stride
````
